```python
import pandas as pd
import numpy as np
import os
```

```python
import pydicom as pdicom
```

```python
list_images = os.listdir('train_images/')
```

```python
data = []
for image in list_images:
    name = image.replace('.dcm', '')
    data_row_img_data = pdicom.read_file('train_images/{}'.format(image))
    x = [name, data_row_img_data.Modality , data_row_img_data.PatientAge , data_row_img_data.PatientSex, data_row_img_data.BodyPartExamined, data_row_img_data.ViewPosition,
         data_row_img_data.ConversionType, data_row_img_data.Rows, data_row_img_data.Columns, data_row_img_data.PixelSpacing]
    data.append(x)
```

```python
df = pd.DataFrame(data)
df.columns = ['patientId', 'Modality', 'PatientAge', 'PatientSex','BodyPartExamined', 'ViewPosition', 'ConversionType','Rows','Columns','PixelSpacing']
df
```
